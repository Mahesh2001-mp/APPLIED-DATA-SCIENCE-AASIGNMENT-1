```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df=pd.read_csv('/content/retail_sales_dataset.csv')
```

```python
# Displaying the info of the dataset
df.info()
```

```python
# First few rows of the dataset
df.head()
```

```python
# statistical analysis of the dataset
df.describe()
```

```python
# checking for number of unique values in each column
df.nunique()
```

```python
# Cheking for null values in the dataset
df.isnull().sum()
```

```python
# Plot 1 Function to generate a bar chart for the product category distribution
def plot_product_category_distribution():
    # Creating a bar chart to display the frequency of each product category
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='Product Category', palette='viridis')
    plt.title('Product Category Distribution')
    plt.xlabel('Product Category')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Plot 2 Function to generate a scatter plot to analyze the relationship between Age and Total Amount
def plot_age_vs_total_amount():
    # Creating a scatter plot for Age vs. Total Amount
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x='Age', y='Total Amount', hue='Gender', palette='coolwarm')
    plt.title('Age vs Total Amount Spent')
    plt.xlabel('Age')
    plt.ylabel('Total Amount')
    plt.tight_layout()
    plt.show()
```

```python
# Plot 3 Function to generate a heatmap for the correlation matrix
def plot_correlation_heatmap():
    # Calculating the correlation matrix
    corr = df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']].corr()
    # Creating a heatmap to visualize correlations
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()
```

```python
# Analyzing relationships and statistics
def analyze_data():
    # Displaying summary statistics
    print("Descriptive Statistics:\n", df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']].describe())
    # Displaying correlation matrix for numeric variables
    print("\nCorrelation Matrix:\n", df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']].corr())
```

```python
# Analyzing and  the functions to generate plots and display statistics
analyze_data()
```
